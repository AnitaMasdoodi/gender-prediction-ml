```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3>Creating a dataframe</h3>

```python
import os
df = pd.read_csv(os.path.join("Data.csv"))
df.head(9)
```

<h3>Displaying the numbers of row & column</h3>

```python
df.shape
```

<h3>Diagram</h3>

```python
viz = df[["Hight", "Weight", "Shoe_size"]]
viz.hist()
plt.show()
```

```python
sns.scatterplot(data=df, x="Hight", y="Weight", hue="Gender", palette={"Male": "blue", "Female": "red"})
plt.title("Height vs Weight by Gender")
plt.show()
```

<h3>Pre-processing</h3>

```python
x = df[["Hight", "Weight", "Shoe_size"]].values
x
```

```python
y = df["Gender"].values
y
```

<h3>Creating the train & test array</h3>

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

<h3>Modeling and fit</h3>

```python
from sklearn.tree import DecisionTreeClassifier
cdf = DecisionTreeClassifier()
cdf.fit(x_train, y_train)
```

<h3>Prediction</h3>

```python
y_hat = cdf.predict(x_test)
print(x_test)
print(y_hat)
print(y_test)
```


<h3>Evaluation</h3>



```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_hat))
```
